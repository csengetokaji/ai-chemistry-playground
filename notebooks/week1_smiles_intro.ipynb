```python
from rdkit import Chem
from rdkit.Chem import Draw, Descriptors

# 5 SMILES molekula
smiles_list = [
    'CC(=O)OC1=CC=CC=C1C(=O)O',  # Aspirin
    'CCN1C=NC2=C1C(=O)N(C(=O)N2C)C',  # Caffeine
    'CN1C=NC2=C1C(=O)N(C(=O)N2C)C',  # Theobromine
    'C1=CC=C(C=C1)C=O',  # Benzaldehyde
    'CC(C)CC1=CC=C(C=C1)C(C)C(=O)O'  # Ibuprofen
]

# Molekulák beolvasása és rajzolása
mols = [Chem.MolFromSmiles(smi) for smi in smiles_list]
Draw.MolsToGridImage(mols, molsPerRow=3)

# Tulajdonságok kiírása
for i, mol in enumerate(mols):
    name = f"Molecule {i+1}"
    mw = Descriptors.MolWt(mol)
    logp = Descriptors.MolLogP(mol)
    print(f"{name}: MolWt = {mw:.2f}, logP = {logp:.2f}")
```

# Molekulák vizsgálata
Ez a notebook 5 molekula SMILES-alapú feldolgozását végzi RDKit segítségével.

**Tulajdonságok:** moláris tömeg, logP érték.


```python
names = ['Aspirin', 'Caffeine', 'Theobromine', 'Benzaldehyde', 'Ibuprofen']
mols = [Chem.MolFromSmiles(smi) for smi in smiles_list]
Draw.MolsToGridImage(mols, legends=names, molsPerRow=3)
```
